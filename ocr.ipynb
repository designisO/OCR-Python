```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Read in images or video
IMAGE_PATH = "chairmotto.jpg"
```

```python
reader = easyocr.Reader(['en'], gpu=False)
result = reader.readtext(IMAGE_PATH)
result
```

```python
# Draw Results
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][0])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
# visulalization of the optical characters (OC) on the image.

img = cv2.imread(IMAGE_PATH) # read img

img = cv2.rectangle(img, top_left, bottom_right, (0,255,0), 5) # draw rectangle
img = cv2. putText(img, text, top_left, font, .5, (255, 255, 255), 2, cv2.LINE_AA) # place text on the image
plt.imshow(img)
plt.show()
```

```python

img = cv2.imread(IMAGE_PATH)
for detection in result:
    top_left = tuple([int(val) for val in detection[0][0]])
    bottom_right = tuple([int(val) for val in detection[0][0]])
    font = cv2.FONT_HERSHEY_SIMPLEX
    img = cv2.rectangle(img, top_left, bottom_right, (0,255,0), 5) # draw rectangle
    img = cv2. putText(img, text, top_left, font, .5, (255, 255, 255), 2, cv2.LINE_AA) # place text on the image

plt.figure(figsize=(10,10))
plt.imshow(img)
plt.show()
```
